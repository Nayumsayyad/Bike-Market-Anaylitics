```python
Q1 = df['on_road_price_inr'].quantile(0.25)
Q3 = df['on_road_price_inr'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df['on_road_price_inr'] < lower) |
    (df['on_road_price_inr'] > upper)
]

outliers
```

```python
# gst impact percentage
df['gst_impact_pct'] = (
    df['gst_amount_inr'] /
    df['factory_price_inr']
) * 100
```

```python
# Which segment pays highest tax
sns.barplot(
    x='segment',
    y='gst_impact_pct',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

```python
# Which buyers are most price sensitive
sns.boxplot(
    x='buyer_behaviour',
    y='price_sensitivity',
    data=df
)

plt.show()
```

```python
df['price_difference'] = (
    df['on_road_price_inr'] -
    df['factory_price_inr']
)
```

```python
df['value_score'] = (
    df['mileage_kmpl'] *
    df['overall_score']
) / df['on_road_price_inr']
```
